Detection Anomaly

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
path_peserta_bpjs = "data/db/jumlah warga terdaftar peserta bpjs kesehatan.csv"
df = pd.read_csv(path_peserta_bpjs)
df['tahun'] = df['tahun'].astype(int)
```

```python
df.head()
```

```python
def generate_trend_peserta(selected_kota: list[str]) -> pd.DataFrame:
    if selected_kota:
        df_filtered = df[df['bps_nama_kabupaten_kota'].isin(selected_kota)]
    else:
        df_filtered = df.copy()

    df_agg = (
        df_filtered
        .groupby(['tahun', 'bps_nama_kabupaten_kota'])['jumlah_ warga_terdaftar_bpjs']
        .sum()
        .reset_index()
        .sort_values(by='tahun')
    )

    return df_agg
```

```python
df_agg = generate_trend_peserta(["KABUPATEN GARUT", "KABUPATEN INDRAMAYU"])
```

```python
df_agg
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import plotly.express as px


# Fungsi forecasting
def forecast_linear(df, tahun_pred=2026):
    pred_all = pd.DataFrame()
    
    for wilayah in df["bps_nama_kabupaten_kota"].unique():
        df_w = df[df["bps_nama_kabupaten_kota"] == wilayah]
        
        X = df_w["tahun"].values.reshape(-1, 1)
        y = df_w["jumlah_ warga_terdaftar_bpjs"].values
        
        model = LinearRegression()
        model.fit(X, y)
        
        tahun_future = np.arange(df_w["tahun"].min(), tahun_pred + 1).reshape(-1, 1)
        y_pred = model.predict(tahun_future)
        
        pred_df = pd.DataFrame({
            "tahun": tahun_future.flatten(),
            "prediksi": y_pred.astype(int),
            "bps_nama_kabupaten_kota": wilayah
        })
        
        pred_all = pd.concat([pred_all, pred_df], ignore_index=True)
        
    return pred_all

# Jalankan forecasting
```

```python
pred_df = forecast_linear(df_agg, 2026)
```

```python
pred_df
```

```python
import plotly.io as pio
pio.renderers.default = "notebook_connected"
fig.show()
```

```python

# Plotly chart
fig = px.line(
    pred_df,
    x="tahun",
    y="prediksi",
    color="bps_nama_kabupaten_kota",
    markers=True,
    title="Forecasting Linear Regression Jumlah Peserta BPJS per Kabupaten",
    labels={
        "tahun": "Tahun",
        "prediksi": "Jumlah Peserta BPJS",
        "bps_nama_kabupaten_kota": "Kabupaten/Kota"
    }
)

# Tambahkan data aktual
for wilayah in df_agg["bps_nama_kabupaten_kota"].unique():
    df_w = df_agg[df_agg["bps_nama_kabupaten_kota"] == wilayah]
    fig.add_scatter(
        x=df_w["tahun"],
        y=df_w["jumlah_ warga_terdaftar_bpjs"],
        mode="markers+lines",
        name=f"Actual {wilayah}",
        line=dict(dash="solid")
    )

fig.show()
```

# Prophet forecasting 

```python
import pandas as pd
import numpy as np
from prophet import Prophet
import plotly.express as px
```

```python
df = generate_trend_peserta(["KABUPATEN GARUT", "KABUPATEN INDRAMAYU"])
```

```python
df["ds"] = pd.to_datetime(df["tahun"].astype(str) + "-01-01")
df.rename(columns={"jumlah_ warga_terdaftar_bpjs": "y"}, inplace=True)

# Forecast Prophet per wilayah
pred_all = pd.DataFrame()

for wilayah in df["bps_nama_kabupaten_kota"].unique():
    df_w = df[df["bps_nama_kabupaten_kota"] == wilayah][["ds", "y"]].copy()
    
    model = Prophet(yearly_seasonality=False, daily_seasonality=False)
    model.fit(df_w)
    
    # Buat future dataframe sampai 2026
    future = model.make_future_dataframe(periods=3, freq="Y")  # 3 tahun ke depan
    forecast = model.predict(future)
    
    forecast["bps_nama_kabupaten_kota"] = wilayah
    pred_all = pd.concat([pred_all, forecast[["ds", "yhat", "bps_nama_kabupaten_kota"]]], ignore_index=True)

```

```python
# Plotly chart
fig = px.line(
    pred_all,
    x="ds",
    y="yhat",
    color="bps_nama_kabupaten_kota",
    markers=True,
    title="Forecasting Jumlah Peserta BPJS dengan Prophet",
    labels={
        "ds": "Tahun",
        "yhat": "Jumlah Peserta BPJS (Prediksi)",
        "bps_nama_kabupaten_kota": "Kabupaten/Kota"
    }
)

# Tambahkan data aktual
for wilayah in df["bps_nama_kabupaten_kota"].unique():
    df_w = df[df["bps_nama_kabupaten_kota"] == wilayah]
    fig.add_scatter(
        x=df_w["ds"],
        y=df_w["y"],
        mode="markers+lines",
        name=f"Actual {wilayah}",
        line=dict(dash="solid")
    )

fig.show()
```

ARIMA

```python
path_peserta_bpjs = "data/db/jumlah warga terdaftar peserta bpjs kesehatan.csv"
df = pd.read_csv(path_peserta_bpjs)
df['tahun'] = df['tahun'].astype(int)
```

```python
df = generate_trend_peserta(["KABUPATEN GARUT", "KABUPATEN INDRAMAYU"])
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import plotly.express as px


df["ds"] = pd.to_datetime(df["tahun"].astype(str) + "-01-01")
df.rename(columns={"jumlah_warga_terdaftar_bpjs": "y"}, inplace=True)

# Forecast ARIMA per wilayah
pred_all = pd.DataFrame()

for wilayah in df["bps_nama_kabupaten_kota"].unique():
    df_w = df[df["bps_nama_kabupaten_kota"] == wilayah].sort_values("ds")
    y = df_w["y"].values
    
    # ARIMA(1,1,1) sederhana
    model = ARIMA(y, order=(1, 1, 1))
    model_fit = model.fit()
    
    # Prediksi sampai 2026
    steps_ahead = max(1, 2026 - df_w["ds"].dt.year.max())
    forecast = model_fit.forecast(steps=steps_ahead)
    
    # Gabung hasil prediksi ke dataframe
    future_years = pd.date_range(start=df_w["ds"].max() + pd.DateOffset(years=1), periods=steps_ahead, freq="Y")
    df_pred = pd.DataFrame({
        "ds": future_years,
        "yhat": forecast.values,
        "bps_nama_kabupaten_kota": wilayah
    })
    
    pred_all = pd.concat([pred_all, df_pred], ignore_index=True)


```

```python
# Gabungkan data aktual untuk plot
fig = px.line(
    pred_all,
    x="ds",
    y="yhat",
    color="bps_nama_kabupaten_kota",
    markers=True,
    title="Forecasting Jumlah Peserta BPJS dengan ARIMA",
    labels={
        "ds": "Tahun",
        "yhat": "Jumlah Peserta BPJS (Prediksi)",
        "bps_nama_kabupaten_kota": "Kabupaten/Kota"
    }
)

for wilayah in df["bps_nama_kabupaten_kota"].unique():
    df_w = df[df["bps_nama_kabupaten_kota"] == wilayah]
    fig.add_scatter(
        x=df_w["ds"],
        y=df_w["y"],
        mode="markers+lines",
        name=f"Actual {wilayah}",
        line=dict(dash="solid")
    )

fig.show()
```
